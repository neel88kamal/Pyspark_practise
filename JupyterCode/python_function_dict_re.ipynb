```python
'''
The prices are stored in a list
              prices = [1.09, 23.56, 57.84, 4.56, 6.78]
              tax_rate = 0.08
              calculate the final price where the final_price is price + price * tax_rate
              
'''
def final_price(price,tax_rate):
    return round(price + (price * tax_rate),2)
    
prices = [1.09, 23.56, 57.84, 4.56, 6.78]
tax_rate = 0.08

final_price_list = [final_price(x,tax_rate) for x in prices]
print(final_price_list)
```

```python
#First bubble sort a list then we will try dictionary
list11=[22,64,35,11,13,90]
n = len(list11)
for i in range(n-1):
    for j in range(0,n-i-1):
        if list11[j]>list11[j+1]:
            list11[j],list11[j+1] = list11[j+1],list11[j]
print(list11)
```

```python
#Now lets sort dictionary by doing bubble sort i.e sort it without using sorted function
dictionary_1 = {'student1': 45, 'student2': 75, 'student3': 65, 'student4': 85}
print(dictionary_1.items())
n = len(dictionary_1.items())
items= list(dictionary_1.items())
for i in range(n):
    for j in range(0,n-i-1):    #this range of n-i-1 is specifically done to do efficient comparison, actually we could have gone from (0,n) as well,
        #but in every one iteration biggest item gets sorted automatically to the end of the list. with i=0 we go till n-0-1, but with i=1, we only go till
        #n-1-1, i.e we compare only from 0 to n-2 i.e one element less and next time i=2, again only n-3 bcoz last element is getting sorted at right place
        #automatically
        if items[j][1]> items[j+1][1]:
            items[j],items[j+1]=items[j+1],items[j]
print(items)
```
