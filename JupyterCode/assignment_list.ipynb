```python
#Question: Find the index of the last occurrence of an element in a list.
list1=[1,2,1,2,2,2,3,4]
sort_list=sorted(list1)
element_to_search=2
last_index = 0
for i in range(len(sort_list)):
    if sort_list[i]==element_to_search:
        last_index = i
print(f'last occurance of {element_to_search} in {sort_list} is at {last_index}')
```

```python
#Question: write a function that find the difference between two lists.
list_23=[1,2,3,4,5]
list_33=[2,3,4,99]
diff_list=[]
for i in list_23:
    if i not in list_33:
        diff_list.append(i)
print(diff_list)

#can we write list comprehension instead of for loop
diff_list_1=[i for i in list_23 if i not in list_33]
print(diff_list_1)

#THere are other ways to do this as well
#first method:
difference = list(set(list_23).difference(set(list_33)))
print(difference)
#2nd method:
difference = list(set(list_23)-(set(list_33)))
print(difference)

#what if we need to find elements which are not common of each list, will we run two loop one after other or there could be better way than this.
difference_total = list(set(list_23)-(set(list_33))) + list(set(list_33)-set(list_23))
print(difference_total)
```

```python
#Question: write a function that calculate the product of all elements in the list.
def product_list(list91):
    product=1
    for i in range(len(list91)):
        product=product*int(list91[i])
    return product
print(product_list([2,3,4,5]))

#there could be one more way, slight modification to get this done
def product_list(list91):
    product=1
    for i in list91:
        product=product*int(i)
    return product
print(product_list([2,3,4,5]))
```

```python
#Question: write a funciton that will Flatten a list of lists into a single list.
list_of_list1 = [[1,2],[3,4],[5,6,7],[3,9]]
flat_list=[]
for i in list_of_list1:
    for j in i:
        flat_list.append(j)
print(flat_list)
#if we need to find unique list, then we can do set at the end
print(list(set(flat_list)))
```

```python
#can we use recursion to print febonachi series upto a number
#0,1,1,2,3,5,8,13
#question is print febonachi series upto 30
num=30
final_feb_series=[]
def febonachi_series(a,b):
    if a>num:
        return 1
    print(a)
    final_feb_series.append(a)
    return febonachi_series(b,a+b)

febonachi_series(0,1)
print(final_feb_series)


#Method-2: to do same thing
def fibonacci_recursive(n, a=0, b=1):
    if a > n:
        return
    print(a, end=" ")
    fibonacci_recursive(n, b, a + b)
# Example usage
fibonacci_recursive(30)


#method-3 to do same thing
print("\n")
def fibonacci_recursive_while(n):
    a,b=0,1
    while (a<n):
        print(a,end=" ")
        a,b = b, a+b

fibonacci_recursive_while(30)
```
