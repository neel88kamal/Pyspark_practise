```python
# count vowels in a given word word
word1 = "hEllo how arE you"
vowels =['a','e','i','o','u']
count = 0
for i in word1:
    if i.lower() in vowels:
        count+=1
#using list comprehension
final_val = sum(1 for i in word1 if i.lower() in vowels)
print(count)
print(final_val)
```

```python
#write a function that will create a fibonacci series
#print febo series upto number n
def febonachi(n):
    a=0
    b=1
    sum=0
    febo_series=[a,b]
    while sum<n:
        sum = a+b
        if sum>n:
            break
        else:
            febo_series.append(sum)
            a=b
            b=sum
    return febo_series

febonachi(30)
```

```python
#Given the names and grades for each student in a class of students, 
#store them in a nested list and print the name(s) of any student(s) having the second lowest grade.
student = [['ram', 60],['shyam', 70],['Hira',45],['Jiru',90]]
#sort it basis of score, and then pick 2nd lowest
for i in range(len(student)):
    for j in range(len(student)-1):
        if student[j][1] > student[j+1][1]:
            student[j], student[j+1] = student[j+1], student[j]
        
print(student)
print(f'student having 2nd lowest grade is {student[1][0]}')
```

```python
#You have been given 3 integers - l, r and k. Find how many numbers between l and r (both inclusive) are divisible by k. 
#You do not need to print these numbers, you just have to find their count.
def find_divisible_number(l,r,k):
    return sum(1 for i in range(l,r+1) if i%k==0 )
print(find_divisible_number(20,30,10))
```

```python
#Get only the numbers in a sentence like Input = 'In 1984 there were 13 instances of a protest with over 1000 people attending'. 
#result = [1984, 13, 1000]
import re
Input = 'In 1984 there were 13 instances of a protest with over 1000 people attending'
pattern = re.compile(r'\d+')
matches = pattern.finditer(Input)
result=[]
for match in matches:
    result.append(match.group(0))
print(result)
```
