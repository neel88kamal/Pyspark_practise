```python
whatiamlookingfor=soup.find_all("div", class_="card")
# print(whatiamlookingfor)
courses=[]
fees=[]
for i in whatiamlookingfor:
    oneblock= i.find_all("div", class_="card-body")
    for one in oneblock:
        course=one.find("h5").text
        fee = one.find("a").text
        # print(course)
        # print(fee)
        courses.append(course)
        fees.append(fee)
print(courses)
print(fees)
for fee in fees:
    pattern = re.compile(r'\d+\$')
    matches = pattern.findall(fee)  # Changed to 'findall' instead of 'find_all'
    print(matches)   
```
