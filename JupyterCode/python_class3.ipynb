```python
def area_circle(rad):
    return 22/7*rad**2

radiuses =[10,20,30,40,50]
areas=[]

for i in radiuses:
    areas.append(round(area_circle(i),3))
print(areas)
```

```python
#can we do above using list comprehension
radiuses =[10,20,30,40,50]
final_list= [round(22/7* x **2, for x in radiuses]
```

```python
print(final_list)
```

```python
radiuses =[10,20,30,40,50]
final_list1= [round(area_circle(x),3) for x in radiuses if x%4==0]
print(final_list1)
```
