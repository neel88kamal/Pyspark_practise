```python
farenheit = [120, 45, 60, 88, 101, 44]
for temp in farenheit:
    celcius = (temp - 32) * 5/9
    print("for farenheit ",temp, " equivalent celcius tempreature is",round(celcius,3),"\n")
```

```python
number = 17
#to find out if this is a prime number
is_prime=True
for num in range(2,number):
    if number % num ==0:
        is_prime=False
        break
    else:
        continue
if is_prime==True:
    print(number," is prime")
else:
    print(number," is not prime")
```

```python
#given a number, use loop to find the factors of the number
number =8
message=""
for num in range(1,number+1):
    if number % num==0:
        message= message+","+str(num)
print("factors are ", message)
```

```python
#given a number, use loop to find the factorial of the number
#factorial of 3 is 3*2*1 = 6
number = int(input("enter a number between 1 to 20: "))
factorial =1
for num in range(1, number+1):
    factorial = factorial*num
print("factorial is ", factorial)
```
